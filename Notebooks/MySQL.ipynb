```python
import pandas as pd
import numpy as np
from getpass import getpass
import pymysql
import sqlalchemy as alch
import cryptography
```

```python
#I load my CSV that ill be using to create my DataBase

df_movies_limpio = pd.read_csv("../Data/df_movies_limpio.csv")

#I hide my password using getpass

contraseña = getpass("Introduce tu contraseña:")

#I create my conexion to the DataBase

bbdd = "proyecto_etl"
conexion = f"mysql+pymysql://root:{contraseña}@localhost/{bbdd}"
engine = alch.create_engine(conexion)
```

```python
#I have one Nan value in this DataFrame, i fill it so i dont have to worry about it later.

df_movies_limpio["Autor_libro"].fillna("No info", inplace=True)
df_movies_limpio["Nota_promedio_libro"].fillna(0, inplace=True)
df_movies_limpio["Votos_libro"].fillna(0, inplace=True)
```

```python
#Here i prepare all of the IDs that i need to create my DataBase.

df_directores = pd.DataFrame(df_movies_limpio["Director"].unique()).reset_index()
df_directores.rename(columns={"index":"ID_Directores"}, inplace = True)

df_autores = pd.DataFrame(df_movies_limpio["Autor_libro"].unique()).reset_index()

df_autores.rename(columns={"index":"ID_Autores"}, inplace = True)

df_movies_limpio = pd.DataFrame(df_movies_limpio).reset_index()
df_movies_limpio.rename(columns ={"index":"ID_Libros"}, inplace = True )
df_movies_limpio = pd.DataFrame(df_movies_limpio).reset_index()
df_movies_limpio.rename(columns ={"index":"ID_Rating"}, inplace = True )
df_movies_limpio.reset_index(inplace = True)
df_movies_limpio.rename(columns={"index":"ID_Peliculas"}, inplace = True )
df_movies_limpio = pd.merge(df_directores,df_movies_limpio, how = "inner", left_on = 0, right_on = "Director")

df_movies_limpio = pd.merge(df_movies_limpio,df_autores, how = "inner", left_on = "Autor_libro",right_on = 0)
```

```python
#Here im replacing all of the characters that may be conflictive in MySQL.

df_movies_limpio["Series_Title"] = df_movies_limpio["Series_Title"].apply(lambda x: x.replace("'",""))
df_movies_limpio["Series_Title"] = df_movies_limpio["Series_Title"].apply(lambda x: x.replace('"',""))
df_movies_limpio["Overview"] = df_movies_limpio["Overview"].apply(lambda x: x.replace("'",""))
df_movies_limpio["Overview"] = df_movies_limpio["Overview"].apply(lambda x: x.replace('"',""))
```

```python
#This is the loop that inserts the Director table.

for index, row in df_directores.iterrows():
        idDirector = engine.execute(f"""select idDirector from director WHERE idDirector = "{row['ID_Directores']}";""")

        if len(list(idDirector)) > 0:
                print("Esta id ya existe")
        else:
                try:
                        engine.execute(f"""
                                INSERT INTO Director (idDirector,Director) VALUES
                                ("{row['ID_Directores']}","{row[0]}");""")

                except:
                        print("no chuta")
                        pass
```

```python
#This is the loop that inserts the Rating table.

for index, row in df_movies_limpio.iterrows():
        idRating = engine.execute(f"""select idRating from rating WHERE idRating = "{row['ID_Rating']}";""")

        if len(list(idRating)) > 0:
                print("Esta id ya existe")
        else:

                try:
                        engine.execute(f"""
                                INSERT INTO rating (idRating, IMDB_rating, Votos_IMDB) VALUES
                                ("{row['ID_Rating']}","{row['IMDB_Rating']}", "{row['No_of_Votes']}");""")

                except:
                        print("no chuta")
                        pass
```

```python
#This is the loop that inserts the Author table.

for index, row in df_autores.iterrows():
        
        idAutores = engine.execute(f"""select idAutor from Autor WHERE idAutor = "{row['ID_Autores']}";""")

        if len(list(idAutores)) > 0:
                print("Esta id ya existe")
        else:
                try:
                        engine.execute(f"""
                                INSERT INTO Autor (idAutor, Autor_libro) VALUES
                                ("{row['ID_Autores']}","{row[0]}");""")

                except:
                        print("no chuta")
                        pass
```

```python
#This is the loop that inserts the Books table.

for index, row in df_movies_limpio.iterrows():
        
        idLibros = engine.execute(f"""select idLibros from libros WHERE idLibros = "{row['ID_Libros']}";""")

        if len(list(idLibros)) > 0:
                print("Esta id ya existe")
        else:
                try:
                        engine.execute(f"""
                                INSERT INTO libros (idLibros, Nota_promedio_libro, Votos_libro, Autor_idAutor) VALUES
                                ("{row['ID_Libros']}","{row['Nota_promedio_libro']}","{row['Votos_libro']}", "{row['ID_Autores']}");""")

                except:
                        print("no chuta")
                        pass
```

```python
#This is the loop that inserts the Movie table.

for index, row in df_movies_limpio.iterrows():
        
        idPeliculas = engine.execute(f"""select idPeliculas from peliculas WHERE idPeliculas = "{row['ID_Peliculas']}";""")

        if len(list(idPeliculas)) > 0:
                print("Esta id ya existe")
        else:
                try:
                        engine.execute(f"""
                                INSERT INTO peliculas (idPeliculas, Series_Title, Released_Year, Runtime, Genre, Overview, rating_idRating, libros_idLibros, Director_idDirector) VALUES
                                ("{row['ID_Peliculas']}","{row['Series_Title']}", "{row['Released_Year']}", "{row['Runtime']}", "{row['Genre']}", "{row['Overview']}","{row['ID_Rating']}","{row['ID_Libros']}","{row['ID_Directores']}");""")
                
                except:
                        print(index)
                        pass
```
